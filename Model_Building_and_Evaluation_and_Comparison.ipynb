```python
lag_features = ['Births', 'cpi index', 'UnemploymentRate', 'Maternities', 'Live Births',
                'BMI (Obesity %)', 'Diabetes %', 'Induction %', 'Method of Birth',
                'Preterm%', 'FulltTerm%', 'Neonatal mortality %', 'SGA%', 'AGA%',
                'LGA%', 'total', 'female%']

for col in lag_features:
    df[f'{col}_lag1'] = df[col].shift(1)
    df[f'{col}_lag3'] = df[col].shift(3)
    df[f'{col}_lag6'] = df[col].shift(6)
df.dropna(inplace=True)

display(df.head())
display(df.tail())
```

**VAR MODEL**

```python
import pandas as pd
from statsmodels.tsa.api import VAR
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = df[['Births', 'UnemploymentRate']].copy()
data = data.dropna().sort_index()
```

```python
data = data.loc[:, data.nunique() > 1]
```

```python
train = data[:'2020-12-01']
test = data['2021-01-01':]
```

```python
model = VAR(train)
var_model = model.fit(2)
```

```python
forecast = var_model.forecast(train.values[-2:], steps=len(test))
forecast_df = pd.DataFrame(forecast, index=test.index, columns=data.columns)
```

```python
y_true = test['Births']
y_pred = forecast_df['Births']
```

```python
var_mae = mean_absolute_error(y_true, y_pred)
var_rmse = np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
print(f"VAR Model MAE: {var_mae:.2f}")
print(f"VAR Model RMSE: {var_rmse:.2f}")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_true, label='Actual Births')
plt.plot(y_pred, label='Predicted Births', linestyle='--')
plt.title('VAR Model Forecast (Births)')
plt.xlabel('Date')
plt.ylabel('Births')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
selected = ['Births', 'UnemploymentRate', 'cpi index']
data = df[selected].dropna().sort_index()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(data)

# Function to create input sequences for LSTM
def create_sequences(data, window_size):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i - window_size:i, :-1])  # input features
        y.append(data[i, -1])  # target (Births)
    return np.array(X), np.array(y)

window_size = 6
X_seq, y_seq = create_sequences(scaled, window_size)

```

```python
split = int(0.8 * len(X_seq))
X_train, X_test = X_seq[:split], X_seq[split:]
y_train, y_test = y_seq[:split], y_seq[split:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

model = Sequential([
    LSTM(64, activation='tanh', input_shape=(X_train.shape[1], X_train.shape[2])),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=30, batch_size=16, verbose=1)
```

```python
y_pred = model.predict(X_test)

y_pred_orig = scaler.inverse_transform(np.concatenate((np.zeros((len(y_pred), 2)), y_pred), axis=1))[:, -1]
y_test_orig = scaler.inverse_transform(np.concatenate((np.zeros((len(y_test), 2)), y_test.reshape(-1,1)), axis=1))[:, -1]
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

lstm_mae = mean_absolute_error(y_test_orig, y_pred_orig)
lstm_rmse = np.sqrt(mean_squared_error(y_test_orig, y_pred_orig))
print(f"LSTM MAE: {lstm_mae:.2f}")
print(f"LSTM RMSE: {lstm_rmse:.2f}")

plt.figure(figsize=(6,4))
plt.plot(y_test_orig, label="Actual")
plt.plot(y_pred_orig, label="Predicted", linestyle="--")
plt.title("LSTM Forecast vs Actual")
plt.legend()
plt.show()
```
